```python
def xstar2table_input():
    
# Function to create input file xstar2table.par 

# Remove old xstar2table.* and xout* files and create new xout* files 
    os.system('rm -f ./xstar2table.* xout*')
    os.system('cp xstinitable.fits xout_ain.fits')
    os.system('cp xstinitable.fits xout_aout.fits')
    os.system('cp xstinitable.fits xout_mtable.fits')
    os.system('cp xstinitable.fits xout_etable.fits')

    file=open('./xstar2table.par','a')
# Construct new xstar2table.par file
    xstar2table_in = 'xstarspec,s,h,xout_spect1.fits,,,"XSTAR Spectral File"\n\
atableinfile,s,h,xout_ain.fits,,,"Inward ATABLE file"\n\
atableoutfile,s,h,xout_aout.fits,,,"Outward ATABLE file"\n\
mtablefile,s,h,xout_mtable.fits,,,"MTABLE file"\n\
etablefile,s,h,xout_etable.fits,,,"ETABLE file"\n\
mode,s,h,ql,,,"mode"'    
    file.write(xstar2table_in)
    file.close()
```

```python
def xstarrun(par):
    dir=par.split()[len(par.split())-1]
    dir=dir.replace("'","")
    os.mkdir(dir)
    os.chdir(dir)
    os.system(par)
    os.chdir('..')
    return dir
```

```python
def multixstar(inpar,num_threads=1,trialrun=False,erase=False):

# Maxime: limit number of Xstar runs to num_proc
    num_proc = 16
    
    print('***PyXstar: Begin xstinitable')
    with hsp.utils.local_pfiles_context(os.getcwd()):
        xstinitable_input(inpar)
        hsp_result=hsp.xstinitable(noprompt=True)
        print(hsp_result.stdout)
        f=open('./xstinitable.lis','r')
        lines=f.readlines()
#        num_proc = len(lines)        
        print('***PyXstar: Number of Xstar runs in model grid = ', len(lines))
        print('***PyXstar: Number of threads = ',num_threads)
    if trialrun: return
    
    dirlist = os.listdir('./')
#    print(dirlist)
    i=0
    while i < len(dirlist):
        if 'loop' in dirlist[i]:
            os.system("rm -r " + dirlist[i])
        i += 1
        
    print('***PyXstar: Running ',num_proc,' Xstar models')
    
    pool = multiprocessing.Pool(processes=num_threads)
    for i in range(num_proc):
        pool.apply_async(xstarrun,args=(lines[i],))
    pool.close()
    pool.join()   

    print('***PyXstar: End of Xstar runs. Begin xstar2table')

    with hsp.utils.local_pfiles_context(os.getcwd()):
        xstar2table_input()
        for i in range(num_proc):
            dir = "./loopcontrol="+str(i+1)
            os.chdir(dir)
            os.system("ln -s ../xout_ain.fits .")
            os.system("ln -s ../xout_aout.fits .")
            os.system("ln -s ../xout_mtable.fits .")
            os.system("ln -s ../xout_etable.fits .")
            os.system("ln -s ../xstar2table.par .")
            hsp_result=hsp.xstar2table()
            print(hsp_result.stdout)
            os.chdir("..")  
            
    if erase:      
        dirlist = os.listdir('./')
        i=0
        while i < len(dirlist):
            if 'loop' in dirlist[i]:
                os.system("rm -r " + dirlist[i])
        i += 1
    print('***PyXstar: End of xstar2table')
```
